```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set style
sns.set(style="whitegrid")

# Load the dataset
df = pd.read_csv("data/survey_data.csv")
```

```python
# Sort by total_income and show top N
top_ages = df.sort_values(by="total_income", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x="age", y="total_income", data=top_ages, palette="viridis")
plt.title("Top 10 Ages with Highest Income")
plt.xlabel("Age")
plt.ylabel("Total Income")
plt.tight_layout()

# Save the chart
plt.savefig("charts/ages_highest_income.png")
plt.show()
```

```python
# Melt the DataFrame to long format for seaborn
expense_categories = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
df_long = df.melt(id_vars=["gender"], value_vars=expense_categories, var_name="Category", value_name="Amount")

# Group by gender and category
plt.figure(figsize=(12, 6))
sns.barplot(data=df_long, x="Category", y="Amount", hue="gender", estimator="sum")
plt.title("Gender Distribution of Spending by Category")
plt.ylabel("Total Amount Spent")
plt.xlabel("Expense Category")
plt.legend(title="Gender")
plt.tight_layout()

# Save the chart
plt.savefig("charts/gender_spending_distribution.png")
plt.show()
```
